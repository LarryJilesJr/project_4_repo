```python
# Importing our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Importing and read the charity_data.csv.
import pandas as pd 
heart_disease_df = pd.read_csv("Resources/heart_2022_no_nans.csv")
heart_disease_df.head()
```

```python
# Dropping the non-beneficial ID columns, 'PhysicalHealthDays', 'MentalHealthDays', 'LastCheckupTime', 'SmokerStatus', and 'ECigaretteUsage'.
heart_disease_df.drop(['PhysicalHealthDays', 'MentalHealthDays', 'LastCheckupTime', 'SmokerStatus', 'ECigaretteUsage'], axis=1, inplace=True)
heart_disease_df.head(3) 
```

```python
heart_disease_df.info()
```

```python
# Dropping additional non-beneficial columns, 'HadDepressiveDisorder', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing', and 'DifficultyErrands'
heart_disease_df.drop(['HadDepressiveDisorder', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing', 'DifficultyErrands'], axis=1, inplace=True)

heart_disease_df.head(3) 
```

```python
heart_disease_df.info()
```

```python
# Replacing "Yes" with 1 and "No" with 0 in the specified columns
columns_to_replace = ['HadHeartAttack', 'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer',
                      'HadCOPD', 'HadKidneyDisease', 'HadArthritis', 'HadDiabetes', 'ChestScan',
                      'AlcoholDrinkers', 'HIVTesting', 'FluVaxLast12', 'PneumoVaxEver',
                      'TetanusLast10Tdap', 'HighRiskLastYear', 'CovidPos']

# Replacing specified columns in the DataFrame
heart_disease_df[columns_to_replace] = heart_disease_df[columns_to_replace].replace({'Yes': 1, 'No': 0})

# Showing the DataFrame with replaced values
heart_disease_df.head()
```

```python
# Determining the number of unique values in each column.
unique_values_in_each_column = heart_disease_df.nunique()
unique_values_in_each_column
```

```python
# Looking at 'BMI' value counts for binning
bmi_value_counts = heart_disease_df['BMI'].value_counts()
bmi_value_counts
```

```python
# Looking for 'BMI" value counts greater than 100
bmi_value_counts[bmi_value_counts>100]
```

```python
# Determining which values can be replaced if counts are less than or equal to 100.
names_to_replace = list(bmi_value_counts[bmi_value_counts <= 100].index)

# Replacing values in the DataFrame
for name in names_to_replace:
    heart_disease_df['BMI'] = heart_disease_df['BMI'].replace(name, 'Other')
    
# Checking to make sure the binning process was successful
heart_disease_df['BMI'].value_counts()
```

```python
# Looking at 'State' value counts for binning
state_value_counts = heart_disease_df['State'].value_counts()
state_value_counts
```

```python
# Choose a cutoff value and create a list of state values to be replaced

# Choosing a cutoff value for the number of occurrences
cutoff_value = 5000 

# Creating a list of state value types to be replaced
states_to_replace = state_value_counts[state_value_counts < cutoff_value].index.tolist()

# Replacing in DataFrame
for app in states_to_replace:
    heart_disease_df['State'] = heart_disease_df['State'].replace(app,"Other")

# Checking to make sure binning was successful
heart_disease_df['State'].value_counts()
```

```python
# Looking at 'WeightInKilograms' value counts for binning
weight_in_kilograms_value_counts = heart_disease_df['WeightInKilograms'].value_counts()
weight_in_kilograms_value_counts
```

```python
# Choose a cutoff value and create a list of WeightInKilograms to be replaced

# Choosing a cutoff value for the number of occurrences
cutoff_value = 5000  

# Create a list of WeightInKilograms to be replaced
classifications_to_replace = weight_in_kilograms_value_counts[weight_in_kilograms_value_counts < cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    heart_disease_df['WeightInKilograms'] = heart_disease_df['WeightInKilograms'].replace(cls,"Other")
    
# Check to make sure binning was successful
heart_disease_df['WeightInKilograms'].value_counts()
```

```python
# Generating an array of the DataFrames categorical variables
categorical_variables = list(heart_disease_df.dtypes[heart_disease_df.dtypes == 'object'].index)
categorical_variables
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
new_heart_disease_df = pd.get_dummies(heart_disease_df)

# Displaying the first few rows of the updated DataFrame
new_heart_disease_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = new_heart_disease_df['HadHeartAttack'].values
X = new_heart_disease_df.drop(columns=['HadHeartAttack']).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

# Defining the deep neural net model
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  100
hidden_nodes_layer2 = 30
hidden_nodes_layer3 = 10

nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="sigmoid"))

# Third hidden layer
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="sigmoid"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compiling the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Training the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Exporting our model to HDF5 file

# Defining the filename for saving the model
filename = 'Heart_disease_model.h5'  

# Exporting the model to HDF5 file
nn_model.save(filename)

print(f"Model saved to {filename}")
```
